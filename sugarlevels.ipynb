```python
import pandas as pd

# Membaca file CSV
data = pd.read_csv('dataset.csv')
print(data.head())
```

```python
import matplotlib.pyplot as plt

plt.hist(data['Kadar_Gula'], bins=10, color='skyblue')
plt.title('Distribusi Kadar Gula Darah')
plt.xlabel('Kadar Gula (mg/dL)')
plt.ylabel('Frekuensi')
plt.show()
```

```python
def categorize_gula(x):
    if x < 100:
        return 'Normal'
    elif 100 <= x < 126:
        return 'Pre-diabetes'
    else:
        return 'Diabetes'

data['Kategori'] = data['Kadar_Gula'].apply(categorize_gula)
```
